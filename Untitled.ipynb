```python
# What was the Average 'Quantity Sold' & 'Average Revenue' for each product ?
```

```python
data.head()
```

```python
Pro_Quantity_Revenue = data.groupby('Product').agg({'Quantity':'mean','Revenue':'mean'}).sort_values('Revenue',ascending = False)
Pro_Quantity_Revenue
```

```python
plt.figure(figsize = (10,5));
plt.bar(Pro_Quantity_Revenue.index, Pro_Quantity_Revenue['Revenue'], color='lightgreen',width = 0.3);
plt.title('Product by Average Revenue',color = "Green");
plt.xlabel('Products',color = "Green");
plt.ylabel('Avg_Revenue',color = "Green");
plt.xticks(color = "green");
plt.yticks(color = "green");
```

```python
plt.figure(figsize = (10,5));
plt.bar(Pro_Quantity_Revenue.index, Pro_Quantity_Revenue['Quantity'], color='lightblue',width = 0.3);
plt.title('Product by Average Quantity',color = "Blue");
plt.xlabel('Products',color = "Blue");
plt.ylabel('Avg_Quantity',color = "Blue");
plt.xticks(color = "blue");
plt.yticks(color = "blue");
```
